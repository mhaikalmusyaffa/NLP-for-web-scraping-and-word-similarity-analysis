# IMPORT LIBRARY

```python
# Import library
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from gensim.models import Word2Vec
import pandas as pd
import gensim
import time
import multiprocessing
from datetime import timedelta

nltk.download('punkt')
nltk.download('stopwords')
```

# SCRAPPING MENGGUNAKAN BEAUTIFULSOUP4

```python
# Fungsi untuk melakukan web scraping
def scrape_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    paragraphs = soup.find_all('p')
    text = ' '.join([p.text for p in paragraphs])
    return text

url_to_scrape = 'https://en.wikipedia.org/wiki/History_of_the_Israeli%E2%80%93Palestinian_conflict'
article_text = scrape_website(url_to_scrape)
```

# TOKENISASI DATA

```python

# Preprocessing data
def preprocess_data(text):
    # Tokenisasi kalimat
    sentences = sent_tokenize(text)

    # Tokenisasi kata
    tokenized_words = [word_tokenize(sentence.lower()) for sentence in sentences]

    # Menghapus stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [
        [word for word in sentence if word.isalnum() and word not in stop_words]
        for sentence in tokenized_words
    ]

    return filtered_words

tokenized_data = preprocess_data(article_text)
model = Word2Vec(sentences=tokenized_data, vector_size=300, window=10, min_count=1, workers=8)
model.save('word2vec_model.bin')
```

# TRAINING MODEL

```python
def calculate_similarity_percentage(kata1, kata22):
    similarity_score = model.wv.similarity(kata1, kata2)
    similarity_percentage = (similarity_score + 1) / 2 * 100  # Mengubah nilai ke rentang 0-100
    print(f"Persentase Kemiripan antara '{kata1}' dan '{kata2}': {similarity_percentage} %")

    return similarity_percentage
```

```python


def print_similar_words_with_percentage(model, target_word, similarity_threshold=0.5):
    similar_words = model.wv.most_similar(target_word)
    print(f"Kata-kata yang paling mirip dengan '{target_word}':")
    for word, similarity in similar_words:
        similarity_percentage = (similarity + 1) / 2 * 100  # Konversi ke persentase
        if similarity_percentage >= similarity_threshold:
            print(f"{word}: {similarity_percentage} %")

loaded_model = Word2Vec.load('word2vec_model.bin')
```
